```python
import numpy as np
import matplotlib.pyplot as plt
import math

%config InlineBackend.figure_format = 'retina'


plt.title('$y = x^2 -x +2$')
#phi = np.linspace(0, 2*np.pi, 200)
x = range(-100,100)
#x = r*np.cos(phi)
#y = r*np.sin(phi)
y = [(number**2 - number + 2) for number in x]
plt.annotate('origin', xy =(0,0), xytext = (0,3000), arrowprops={})
plt.plot(x,y)
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.ylim(10500)

plt.savefig('figure1.png')
```

```python
'''
y = sqrt(x)
y = x**3
y = 2**x
y = 1/(x-1)
'''
xx = range(0,100)
yy = [math.sqrt(n) for n in xx]
plt.plot(x,y)
plt.title('$ y = \sqrt{x}$')
plt.ylabel('$\sqrt{x}$')
plt.xlim(0,100)
plt.ylimn(0,10)
plt.grid(True, ls = '--');
```

```python
x3 = range (0,20)
y3 = [n**3 for n in x3]
plt.plot(x3, y3)
plt.title('$ y = x^3$')
plt.ylabel('$x^3$')
plt.xlabel('$z$')
plt.xlim(0,20)
plt.ylim(0)
plt.grid(True, ls = '--');
```

```python
x4 = [x * 0.1 for x in range(0,150)]
y4 = [2** n for n in x4]

plt.plot(x4,y4)

plt.title('$ y = 2^x$')
plt.xlabel(('$x$'))
plt.ylabel('$2^x$')
plt.grid(True, ls = '--')
```

```python
x5 = [x*0.1 for x in range (0,150)]
y5 = [1/(n+1) for n in x5]

plt.plot(x5, y5)
plt.title('$y=1/(x+1)$')

plt.xlabel('$x$')
plt.ylabel('$1/(x+1)$')
plt.grid(True, ls = '--')
plt.grid(True, ls = '--')
plt.show()
```
